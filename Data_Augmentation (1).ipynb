# Working with single image

```python
# Load img
img = image.load_img('/content/cat.151.jpg',target_size =(200,200))
```

```python
# new images formating details

datagen = ImageDataGenerator(
    rotation_range =40, # image rotate degree
    shear_range=0.2,  # 1 ki scaling shift
    zoom_range=0.2,
    horizontal_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2

    # jab image rotate hoge left right jayegi to khali jagag kese fill hogi
    # so there is a -------
    # fill_mode = defult = nearest \\ reflect \\ constant = fill with black
)
```

```python
img = image.img_to_array(img)
```

```python
img
```

```python
input_batch = img.reshape(1,200,200,3)

# we always pass image in batch
# so creat batch of only one image
```

```python
import os
os.makedirs("aug", exist_ok=True)
```

```python
# main infinite image gen code

i =0

#    flow ---> for working with one image
#    flow_from_dir -----> working with batch of images
for outpute in datagen.flow(input_batch,batch_size=1,save_to_dir='aug'):
  i = i+1

  if i==10:
    break
```

# One input ----> Two output

```python
from keras.models import Model
```

```python
model = Model(inputs= x, outputs=[output1,output2])
```

```python
from keras.layers import *
```

```python
x = Input(shape=(3,))

hidden1 = Dense(128,activation='relu')(x)
hidden2 = Dense(64,activation='relu')(hidden1)
```

```python
output1 = Dense(1,activation='sigmoid')(hidden2)
output2 = Dense(1,activation='relu')(hidden2)
```

```python
model.summary()
```

```python
from keras.utils import plot_model
plot_model(model)
```

# Two input ----> One output

```python
model = Model(inputs= [x1,x2], outputs=z1)
```

```python
x1 = Input(shape=(3,))
x2 = Input(shape=(2,))
```

```python
hiddenx1 = Dense(32,activation='relu')(x1)
hiddenx2 = Dense(32,activation='relu')(hiddenx1)
```

```python
hiddenx11 = Dense(32,activation='relu')(x2)
hiddenx22 = Dense(32,activation='relu')(hiddenx11)
hiddenx23 = Dense(32,activation='relu')(hiddenx22)
```

```python
combined = concatenate([hiddenx2, hiddenx23])
```

```python
# apply a FC layer and then a regression prediction on the combined output

z = Dense(2,activation='relu')(combined)
z1 = Dense(1,activation='linear')(z)
```

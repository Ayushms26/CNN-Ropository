```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np
```

```python
model = ResNet50(weights='imagenet')
# hame vahi weigth chahiye jo imageNET model ko train karne par mile the
```

# load image for prediction

```python
# /content/detect_image_3.webp
# /content/detect_image_5.jpg
# /content/image_detect_1.avif
# /content/image_detect_2.jpg
# /content/image_detect_4.jpg
```

```python
img = image.load_img('/content/image_detect_2.jpg', target_size=(224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
```

```python
img
```

# Predict

```python
pred = model.predict(x)
print(decode_predictions(pred, top=3)[0])
```
